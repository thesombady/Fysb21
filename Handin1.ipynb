```python
x = np.linspace(440*6.2, 440*2.03*np.pi, 10000)
y2 = abs(np.sin(440*np.pi-x/2)/(440*np.pi-x/2))**2
```

```python
plt.plot(x, y2,  '.', markersize = 2)
plt.show()
```

```python
eps = 0.001
FWHM = []
for i in range(len(x)):
    if abs(y2[i] - 1/2)<eps:
        print(f"{y2[i]}", x[i])
        FWHM.append(x[i])
Diff = FWHM[1]-FWHM[0]
print(Diff)
```

```python
sigma = Diff/(2*np.sqrt(2*np.log(2)))
```

```python
List = []
import matplotlib.pyplot as plt 
import numpy as np

ylist = 0
T = 3
tau = T/30
x1 = np.linspace(0,T, 10000)
x2 = np.linspace(2*np.pi, 1, 10000)
def y1(x):
    ylist = []
    for i in range(len(x)):
        if x[i]<2*np.pi*tau:
            ylist.append(np.sin(x[i]/tau))
        else:
            ylist.append(0)
    return np.array(ylist)

def fourier(n):
    return -4/(T*tau)*np.sin(4*np.pi**2*tau*n/T)/((1/tau)**2-(2*np.pi*n/T)**2)*np.sin(2*np.pi*x1/T*n)
for i in range(10):
    ylist += fourier(i)
```

```python
plt.plot(x1,y1(x1),'.', markersize = 1, label = "Original")
plt.plot(x1,ylist,'.', markersize = 1, label = "Fourier")
plt.legend()
plt.plot()
```

```python
x = np.linspace(0,5,1000)
def xlist(x):
    y = []
    for i in range(len(x)):
        if x[i] < 1:
            y.append(x[i]**2)
        else:
            l = np.exp(-0.33*x[i])*(np.cos(x[i])+np.sin(x[i]))
            y.append(l)
    return y
plt.plot(x, xlist(x), '.', markersize = 1)
plt.show()
```
