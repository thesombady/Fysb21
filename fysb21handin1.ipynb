```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 5; tau = T/30
```

```python
t = np.linspace(0,T,1000)
```

```python
def p(t):
    xlist = []
    for i in range(len(t)):
        if t[i] <= 2*np.pi*tau:
            xlist.append(np.sin(t[i]/tau))
        else:
            xlist.append(0)
    return np.array(xlist)
```

```python
plt.plot(t,p(t),'.', markersize = 1)
plt.show()
```

```python
def cn(n):
    c = p(t)*np.exp(-1j*2*n*np.pi*t/T)
    return c.sum()/c.size

def fourier(x, iterations):
    f = np.array([2*cn(i)*np.exp(1j*2*i*np.pi*x/T) for i in range(1,iterations+1)])
    return f.sum()
y2 = np.array([fourier(i,20).real for i in t])
plt.plot(t,y2,'.', markersize = 1, label = "Fourier expansion")
plt.plot(t, p(t), markersize = 1, label = "p(t)")
plt.grid()
plt.title("Fourier Spectrum")
plt.show()
```
